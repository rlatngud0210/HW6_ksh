```python
import csv
import matplotlib.pyplot as plt
def main():
    gender=[]
    color = ['blue','skyblue']
    label = ["Male","Female"]
    f = open("2008.csv",'r',encoding = 'ANSI')
    data = csv.reader(f)
    next(data)
    for row in data:
        gender.append(int(row[3]))
        gender.append(int(row[4]))
        break
    plt.title("Percentage of men and women in Jeju in 2008")
    plt.pie(gender,labels = label,autopct='%.2f%%',colors = color)
    plt.show()
    gender=[]
    f = open("2013.csv",'r',encoding = 'ANSI')
    data = csv.reader(f)
    next(data)
    for row in data:
        gender.append(int(row[3]))
        gender.append(int(row[4]))
        break
    plt.title("Percentage of men and women in Jeju in 2013")
    plt.pie(gender,labels = label,autopct='%.2f%%',colors = color)
    plt.show()
    gender=[]
    f = open("2018.csv",'r',encoding = 'ANSI')
    data = csv.reader(f)
    next(data)
    for row in data:
        gender.append(int(row[3]))
        gender.append(int(row[4]))
        break
    plt.title("Percentage of men and women in Jeju in 2018")
    plt.pie(gender,labels = label,autopct='%.2f%%',colors = color)
    plt.show()
    gender=[]
    f = open("2022.csv",'r',encoding = 'ANSI')
    data = csv.reader(f)
    next(data)
    for row in data:
        gender.append(int(row[3]))
        gender.append(int(row[4]))
        break
    plt.title("Percentage of men and women in Jeju in 2022")
    plt.pie(gender,labels = label,autopct='%.2f%%',colors = color)
    plt.show()

if __name__ == '__main__':
    main()
```
