```python
import csv
import matplotlib.pyplot as plt
def add(station,aim,people):
    for i in station:
        if aim == i[0]:
            i[1] += people
            return
    station.append([aim,people])
def main():
    f= open('subway.csv','r',encoding='utf-8')
    boarding_station = []
    gf_station = []
    max_station = []
    xdata = []
    ydata = []
    data = csv.reader(f)
    next(data)
    next(data)
    for row in data:
        add(boarding_station,row[3],int(row[10])+int(row[12]))
        add(gf_station,row[3],int(row[11])+int(row[13]))
        add(max_station,row[3],int(row[10])+int(row[11])+int(row[12])+int(row[13]))
    boarding_station.sort(key=lambda x: x[1])
    busy_boarding = boarding_station[-30:]
    for st in busy_boarding:
        xdata.append(st[0])
    for pp in busy_boarding:
        ydata.append(pp[1])
    plt.rc('font', family='NanumGothic')
    plt.title(" ")
    plt.bar(xdata,ydata)
    plt.xticks(rotation = 'vertical')
    plt.show()
    xdata = []
    ydata = []
    gf_station.sort(key=lambda x:x[1])
    busy_gf = gf_station[-30:]
    for st in busy_gf:
        xdata.append(st[0])
    for pp in busy_gf:
        ydata.append(pp[1])
    plt.title(" ")
    plt.bar(xdata,ydata)
    plt.xticks(rotation = 'vertical')
    plt.show()
    xdata = []
    ydata = []
    max_station.sort(key=lambda x:x[1])
    busy_st = max_station[-30:]
    for st in busy_st:
        xdata.append(st[0])
    for pp in busy_st:
        ydata.append(pp[1])
    plt.title(" ")
    plt.bar(xdata,ydata)
    plt.xticks(rotation = 'vertical')
    plt.show()
if __name__ == '__main__':

    main()
```
