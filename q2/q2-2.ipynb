```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def main():
    f= open('output.csv', 'r', encoding='ANSI') 
    data = csv.reader(f)
    num = []
    for row in range(100):
        num.append(int(next(data)[0]))
        
    counts = np.bincount(num)
    numbers = np.arange(1, len(counts))

    plt.hist(num, bins=6, color='b', align='left', rwidth=0.8, range=[1, 7])

    for x, y in zip(numbers, counts[1:]): 
        plt.text(x, y, y, ha='center', va='bottom')
    plt.title("100 TIMES")
    plt.show()
    
    num = []
    for row in range(1000):
        num.append(int(next(data)[0]))
        
    counts = np.bincount(num)
    numbers = np.arange(1, len(counts))

    plt.hist(num, bins=6, color='b', align='left', rwidth=0.8, range=[1, 7])

    for x, y in zip(numbers, counts[1:]): 
        plt.text(x, y, y, ha='center', va='bottom')
    plt.title("1000 TIMES")
    plt.show()
    
    num=[]
    for row in data:
        num.append(int(row[0]))

    counts = np.bincount(num)
    numbers = np.arange(1, len(counts))

    plt.hist(num, bins=6, color='b', align='left', rwidth=0.8, range=[1, 7])

    for x, y in zip(numbers, counts[1:]): 
        plt.text(x, y, y, ha='center', va='bottom')
    plt.title("100000 TIMES")
    plt.show()
    f.close()

if __name__ == '__main__':
    main()
```
